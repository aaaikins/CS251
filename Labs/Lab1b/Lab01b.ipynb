### 2a. Plot the data in a single plot

Using the [Matplotlib](https://matplotlib.org/) documentation, visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.

1. Import 'lab1.csv' into a ndarray using genfromtxt. 
2. Set figure size to 12x8. Use either [plt.figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) or [rcParams](https://matplotlib.org/3.1.0/tutorials/introductory/customizing.html#matplotlib-rcparams) (*search for `figure.figsize`*).
2. Create a cartesian line graph of last column (on the X axis) vs the 2nd column (Y axis) using a red dashed line. [matplotlib.pyplot introduction](https://matplotlib.org/tutorials/introductory/pyplot.html)
    1. Include a title and axis labels.

```python
import matplotlib.pyplot as plt
lab1 = np.genfromtxt('data/lab1.csv', delimiter=',')
plt.figure(figsize=(12,8))
# print(lab1[:, -1])
# print(lab1[:,1])
plt.plot(lab1[:, -1], lab1[:,1], 'r--')
plt.title('Line Graph of 2nd Column vs Last Column')
plt.xlabel('Last Column')
plt.ylabel('2nd Column')

plt.legend()

plt.show()
```

### 2b. Plot the data in multiple subplots

Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html). 
1. Row 1 is two subplots (a scatterplot and a line graph) of the 2nd column graphed on the x-axis VS the 1st column graphed on the y-axis
2. Row 2 is two subplots (a scatterplot and a line graph) of the 3rd column (x-axis) VS the 2nd column (y-axis)
3. Row 3 is two subplots (a scatterplot and a line graph) of the 3rd column (x-axis) VS the 1st column (y-axis)
4. Include a title and axis labels for each subplot. You can refer to the 1st column as `X`, the 2nd as `Y`, and the 3rd as `Z`.
5. To insure there is enough space between subplots use [subplots_adjust](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
col1 = lab1[:, 0]
col2 = lab1[:, 1]
col3 = lab1[:, 2]

plt.subplot(3, 2, 1)
plt.subplots_adjust(bottom=1, left= 1, right=2, top=2, wspace=1, hspace=1)
plt.scatter(col2, col1)
plt.title("Row 1 Scatterplot")
plt.xlabel("Column 2")
plt.ylabel("Column 1")

plt.subplot(3, 2, 2)
plt.plot(col2, col1)
plt.title("Row 1 Line Graph")
plt.xlabel("Column 2")
plt.ylabel("Column 1")

plt.subplot(3, 2, 3)
plt.scatter(col3, col2)
plt.title("Row 2 Scatterplot")
plt.xlabel("Column 3")
plt.ylabel("Column 2")

plt.subplot(3, 2, 4)
plt.plot(col3, col2)
plt.title("Row 2 Line Graph")
plt.xlabel("Column 3")
plt.ylabel("Column 2")

plt.subplot(3, 2, 5)
plt.scatter(col3, col1)
plt.title("Row 3 Scatterplot")
plt.xlabel("Column 3")
plt.ylabel("Column 1")

plt.subplot(3, 2, 6)
plt.plot(col3, col1)
plt.title("Row 3 Line Graph")
plt.xlabel("Column 3")
plt.ylabel("Column 1")

plt.show()
```
