### 1d. Find the best `k`

- Below, "script" your `predict` method on both spiral datasets 1 and 2. Compute the accuracy on the respective test sets with many different values of `k`.
- Create two well-labeled plots, one for each spiral dataset, showing the accuracy for many different `k` values.

```python
k_values = range(1, 21)

knn1 = KNN(4)
knn1.train(spiral_1_train, spiral_1_train_y)
accuracies_spiral_1 = []

for k1 in k_values:
    y_pred_spiral_1 = knn1.predict(spiral_1_val, k1)
    accuracy_spiral_1 = knn1.accuracy(y=spiral_1_val_y, y_pred=y_pred_spiral_1)
    accuracies_spiral_1.append(accuracy_spiral_1)

knn2 = KNN(4)
knn2.train(spiral_2_train, spiral_2_train_y)
accuracies_spiral_2 = []

for k2 in k_values:
    y_pred_spiral_2 = knn2.predict(spiral_2_val, k2)
    accuracy_spiral_2 = knn2.accuracy(y=spiral_2_val_y, y_pred=y_pred_spiral_2)
    accuracies_spiral_2.append(accuracy_spiral_2)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracies_spiral_1, marker='o', linestyle='-')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k for Spiral Dataset 1')
plt.xticks(np.arange(0, 20, 2)) 

plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracies_spiral_2, marker='o', linestyle='-')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k for Spiral Dataset 2')
plt.xticks(np.arange(0, 20, 2))

plt.tight_layout()
plt.show()
```
